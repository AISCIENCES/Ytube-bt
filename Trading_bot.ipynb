```python
import websocket, json
from binance.exceptions import BinanceAPIException
from binance.client import Client
from binance.enums import *

SYMBOL = "forusdt"
sell_limit, buy_limit = 0.055, 0.045
SOCKET = "wss://stream.binance.com:9443/ws/"+SYMBOL+"@kline_1m"

client = Client(API_KEY, SECRET_KEY)
```

```python
def on_open(ws):
  print("Connction Opened")

def on_close(ws):
  print("Connction close")

def on_error(ws, error):
  print("Here is an error:", error)

def order(side, quantity, symbol="FORUSDT"):
  try:
    order = client.create_order(symbol= symbol, side = side, type= client.ORDER_TYPE_MARKET, quantity = quantity)
  except BinanceAPIException as e:
    print(e)


def on_message(ws, message):
  print("Message Recieved")
  candle_data = json.loads(message)
  closing_rate = float(candle_data['k']['c'])

  print("Closing:", closing_rate)

  if closing_rate >= sell_limit:
    #sell the coins
    print("SELLLL ORDER")
    order("SELL", 1000)
  elif closing_rate <= buy_limit:
    print("BUYYYYYYYYYYY ORDER")
    order("BUY", 200)
    #buy coins

  print("Closing:", closing_rate)
```

```python
ws = websocket.WebSocketApp(SOCKET, on_open=on_open, on_close=on_close, on_error=on_error, on_message=on_message)
ws.run_forever()
```
